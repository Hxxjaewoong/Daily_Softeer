```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = open('./alice.txt').read()
text
```

```python
# WordCloud  
wordcloud = WordCloud().generate(text)
wordcloud
```

```python
# WordCloud      
wordcloud.words_
```

```python
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)   # wordcloud      
plt.axis('off')  #  
plt.show()
```

```python
# max_words:      
wordcloud = WordCloud(max_words=100).generate(text)  #  100 
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)   
plt.axis('off')  
plt.show()
```

```python
from wordcloud import STOPWORDS

STOPWORDS  #   
```

```python
type(STOPWORDS)
```

```python
# add():  
STOPWORDS.add('said')
print(STOPWORDS)
```


# mask  ndarray   

```python
from PIL import Image
import numpy as np


#   ndarray   alice_mask 
alice_mask = np.array(Image.open('./alice_mask.png'))
alice_mask   # 255:  , 0:  
```

```python
wordcloud = WordCloud(max_words=100,
                      stopwords=STOPWORDS,
                      mask=alice_mask,
                      background_color='ivory').generate(text) # 'said' 
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
