## 사전지식
## 기능요구사항
- 대륙 이름이 포함된 문장을 출력하는 함수를 만듭니다.
- 다른 이름이 제공되지 않으면 이 함수는 기본 대륙으로 "아시아"를 사용해야 합니다.
- 스크립트는 기본 대륙으로 함수를 한 번 실행하고, 다른 대륙 이름(아메리카, 유럽, 아프리카)으로 함수를 세 번 실행해야 합니다.
- 스크립트는 종료되기 전에 함수의 모든 인스턴스가 실행을 완료했는지 확인해야 합니다.

## 프로그래밍관점
- Python의 multiprocessing 모듈을 사용하세요.
- 주어진 대륙 이름을 출력하는 함수를 정의하세요.
- Process 클래스를 사용해 이 함수를 호출하여 여러 프로세스를 생성합니다.
- 모든 프로세스가 적절하게 시작되고 완료되었는지 확인하세요.

## 시청결과 및 동작예시
대륙의 이름은 : 아메리카  
대륙의 이름은 : 유럽  
대륙의 이름은 : 아시아  
대륙의 이름은 : 아프리카  

---

```python
from multiprocessing import Process

def print_continent(name="아시아"):
    print(f"대륙의 이름은 : {name}")

if __name__ == "__main__":
    default_process = Process(target=print_continent)
    default_process.start()

    continents = ["아메리카", "유럽", "아프리카"]
    processes = [Process(target=print_continent, args=(continent,)) for continent in continents]

    for process in processes:
        process.start()

    # 모든 프로세스가 완료되기를 기다림
    default_process.join()
    for process in processes:
        process.join()
```
